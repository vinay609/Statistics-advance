```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

low, high, size = 1, 6, 10000  # Simulating a fair 6-sided die


data = np.random.randint(low, high + 1, size)

# Calculate PMF
values, counts = np.unique(data, return_counts=True)
pmf = counts / size

# Plot PMF
plt.bar(values, pmf, width=0.6, color='blue', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution")
plt.show()
```

```python
#3. Calculate the PDF of a Bernoulli distribution
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

p = 0.5
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]

for x, pdf in zip(x_values, pdf_values):
    print(f"P(X={x}) = {pdf}")

```

```python
#6. Calculate and plot the CDF of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low, high = 1, 6
x = np.arange(low, high + 1)
cdf_values = randint.cdf(x, low, high + 1)

plt.step(x, cdf_values, where='mid', color='red', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Discrete Uniform Distribution")
plt.show()
```

```python
#9. Calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Distribution")
plt.show()
```

```python
#14. Calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

x_value = 75
mean, std = 50, 10
z_score = calculate_z_score(x_value, mean, std)
probability = norm.cdf(z_score)

print("Z-score:", z_score)
print("Probability (CDF):", probability)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

sample = np.array([52, 49, 53, 50, 47, 54, 51, 50, 48, 53])
pop_mean, pop_std = 50, 3
sample_mean = np.mean(sample)
sample_size = len(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
#16. Create a confidence interval for a dataset and interpret the result

import numpy as np
import scipy.stats as stats

data = np.random.normal(loc=50, scale=10, size=100)
mean, std = np.mean(data), np.std(data, ddof=1)
conf_interval = stats.norm.interval(0.95, loc=mean, scale=std / np.sqrt(len(data)))

print("95% Confidence Interval:", conf_interval)
```

```python
#21. Compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.normal(50, 5, 1000)
data2 = np.random.normal(50, 15, 1000)

std1, std2 = np.std(data1), np.std(data2)

plt.hist(data1, bins=30, alpha=0.5, label=f"Std Dev: {std1:.2f}", color='blue', edgecolor='black')
plt.hist(data2, bins=30, alpha=0.5, label=f"Std Dev: {std2:.2f}", color='red', edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset from a normal distribution
import numpy as np

data = np.random.normal(50, 10, 1000)
range_value = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

print("Range:", range_value)
print("Interquartile Range (IQR):", iqr)
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 10, 1000)
normalized_data = (data - np.mean(data)) / np.std(data)

plt.hist(normalized_data, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Normalization")
plt.show()
```

```python
#24. Calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

data = np.random.normal(50, 10, 1000)
skewness = skew(data)
kurt = kurtosis(data)

print("Skewness:", skewness)
print("Kurtosis:", kurt)
```
